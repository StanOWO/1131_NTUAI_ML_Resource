### 問題：我是一個幼稚園老師，我想透過各種資訊，讓個性相近的孩子分在同一組，使他們不太會吵架
- [資料集](https://www.kaggle.com/datasets/abtabm/hobby-prediction-basic)

# Environment

```python
# HappyML install
import os

if not os.path.isdir("HappyML"):
  os.system("git clone https://github.com/cnchi/HappyML.git")
```

```python
# Load data
datasetName="Hobby_Data.csv"
datasetName = datasetName.replace(" ", "%20")
if not os.path.isfile(datasetName):
  os.system("wget https://raw.githubusercontent.com/StanOWO/1131_NTUAI_ML_Resource/main/dataset/" + datasetName)
```

# Preprocessing

```python
import HappyML.preprocessor as pp
```

```python
# Read CSV file
dataset = pp.dataset(file="Hobby_Data.csv")
```

```python
# Decomposition
X = dataset
```

```python
# One-Hot Encoding
# import pandas as pd
# import numpy as np
# columns_to_encode = [i for i in range(X.shape[1]) if i not in [5, 6, 12]]
# ary_dummies = pd.get_dummies(X.iloc[:, columns_to_encode])
# column_else=[5,6,12]
# X = np.concatenate((ary_dummies, X.iloc[:, column_else]), axis=1).astype("float64")
X = pp.onehot_encoder(ary=X, columns=[i for i in range(14) if i != 5 or i!=6 or i!=12], remove_trap=True)
```

```python
# Feature Scaling (for PCA Feature Selection)
X = pp.feature_scaling(fit_ary=X, transform_arys=X)
```

# Model(K-means)

```python
# In[] K-Means Clustering
# from sklearn.cluster import KMeans
# kmeans = KMeans(n_clusters=4, init="k-means++", random_state=0)
# Y_pred = kmeans.fit_predict(X)

from HappyML.clustering import KMeansCluster

cluster = KMeansCluster(best_k=4,max_k=10)
Y_pred = cluster.fit(x_ary=X).predict(x_ary=X, y_column="Group number")
```

# Save as csv file

```python
file_name = "group.csv"

dataset = pp.combine(dataset, Y_pred)
dataset.to_csv(file_name, index=False)

print("Generate", file_name,"Successfully")
```
